### 1. Cicles

#### 1.1 While

* The simplest looping mechanism in Python.

```python
# we define a counter
count = 11
# we implement the loop with a while
# the while will continue as long as the condition inside isn't met
while count <= 15:
  print(count)
  count += 1
  if count == 10:
    break
# when the stop condition is met the else is executed
else:
  print("Loop finished")
```

#### 1.2 For

A for loop is used for iterating over a sequence (that is either a list, a tuple, a dictionary, a set, or a string).

```python
# we create a for loop using range
# range creates a list of numbers
for i in range(15):
    if i == 10:
        break
    print(i)
    # when the stop condition is met the else is executed
else:
    print("Loop finished normally.") # This will not be printed
```

```python
# we can provide a start and stop argument to the range list
for x in range(0,5):
  print(x)
# when the stop condition is met the else is executed
else:
  print("Loop finished")
```

```python
# we can use a list instead of range
my_list = range(0,5)
for x in my_list:
  print(x)
```

```python
# we can use a decreasing list
for x in range(5, -5, -1):
  print(x, ", ", end="")
```

```python
# we can change the step size
for x in range(5, -5, -2):
  print(x, ", ", end="")
```

```python
# we can use a list of strings
fruits = ["apple", "banana", "cherry"]
for x in fruits:
  print(x)
```

```python
# we can use a string
for x in "banana":
  print(x)
```

```python
fruits = ["apple", "banana", "cherry"]
for x in fruits:
  if x != "banana":
    print(x)
```

### 2. Higher-order functions

Is a function that fulfills at least one of the following criteria:

* Takes one or more functions as arguments:
  * This allows for the creation of more generic and reusable functions.
  * Example: a sort function could take a comparison function as an argument, allowing it to sort different data types or objects based on various criteria without needing to be rewritten for each specific case.
* Returns a function as its result:
  * This enables the creation of functions that generate other functions.
  * Often used for creating specialized versions of a general function or for implementing techniques like currying or closures.

```python
def aplicar_operacion(func, lista):
  """
 Applies a function to each element of a list..
  """
  return [func(x) for x in lista]

def cuadrado(x):
  return x * x
```

```python
# we create a list of 5 numbers
numbers = [1, 2, 3, 4, 5]
# we pass the function "cuadrado" and the list as arguments for the higher-order function
cuadrados = aplicar_operacion(cuadrado, numbers)
print(cuadrados)
```

### 3. Lambda functions

 * Is a small, anonymous, single-expression function.
 * A lambda function can take any number of arguments, but can only have one expression.
 * Unlike standard functions defined with the def keyword, lambda functions do not require a name and are typically used for short, one-time operations.

Syntax:

lambda arguments : expression

```python
 # we define a simple lambda function with just one argument
 x = lambda a : a + 10
print(x(5))
```

```python
# we define a multiple argument lambda function
x = lambda a, b, c : a + b + c
print(x(5, 6, 2))
```

```python
# we pass a lambda function as argument for the higher-order function
cubos = aplicar_operacion(lambda x: x ** 3, numbers)
print(cubos)
```
