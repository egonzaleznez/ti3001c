### 2. Funciones

* Una función es un grupo de instrucciones autocontenidas dentro del programa que realiza una tarea única.
* Las funciones ayudan al programador a resolver un problema de manera más simple.
* En lugar de escribir un programa que resuelve un problema grande, se escriben varias funciones que resuelven problemas pequeños.
* Variables can be global or local.

```python
# defining a function with a local variable
def localVariable():
  my_variable = 3
  print(my_variable)

localVariable()

my_variable = "Hi"
print(my_variable)
localVariable()
print(my_variable)
```

##### 3. Costo del terreno

```python
def dame_variables():

    # pido variables
    a = int(input("dame a: "))
    b = int(input("dame b: "))
    c = int(input("dame c: "))
    d = int(input("dame d: "))

    return a, b, c, d

def area_rect(base = 1, altura = 1):

    # calcular área del rectángulo
    area = base * altura

    return area

def area_trian(base = 1, altura = 1):

    # calcular área del triángulo
    area = (base * altura)/2

    return area

def costo_total(area1, area2, area3):

    # calcular área total del terreno
    att = area1 - area2 - area3
    # calcular valor del terreno
    vtt = att * 3450

    return vtt

def codigo_costo_terreno():

    # pido variables
    a, b, c, d = dame_variables()
    # calculo las áreas
    a1 = area_rect(a, b)
    a2 = area_trian(b, d)
    a3 = area_trian(b, c)
    # calculo el costo
    vtt = costo_total(a1, a2, a3)

    print("el valor del terreno es: $", vtt)

codigo_costo_terreno()
```

```python
# un programa que pide 2 números y encuentra el mayor
# un ejemplo de cómo implemento if

def buscaMayor():
    x = int(input("dame x: ")) # el dato lo convierto de texto a número
    y = int(input("dame y: "))

    if y < x:
        print("x es mayor")
    elif y == x:
        print("son iguales")
    else:
        print("y es mayor")

#def main():

#    print("Hola")
#    buscaMayor()

buscaMayor()
```

```python
superhero = "Spiderman"
if superhero == "Superman":
  secret_identity = "Clark Kent"
elif superhero == "Batman":
  secret_identity = "Bruce Wayne"
elif superhero == "Spiderman":
  secret_identity = "Peter Parker"
else:
  secret_identity = "?"
print(secret_identity)
```

#### 3.2 Match

* It’s similar to the switch statement in other languages like C and Java.
* You can do the same job with a bunch of if, else, and elif statements.
* It’s sometimes a bit more compact — and, by some benchmarks, a bit faster.

```python
superhero = "Spiderman"
match superhero:
  case "Superman":
    secret_identity = "Clark Kent"
  case "Batman":
    secret_identity = "Bruce Wayne"
  case "Spiderman":
    secret_identity = "Peter Parker"
  # if nothing else matches
  case _:
    secret_identity = "?"
print(secret_identity)
```
